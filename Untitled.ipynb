```python
batsman=pd.read_csv('sharma-kohli.csv')
batsman
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
# labels title
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title('virat vs rohit record')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='solid',linewidth=4)
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='dashdot',linewidth=4)
plt.title('virat vs rohit record')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# marker and markersize
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='solid',linewidth=4,marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='dashdot',linewidth=4,marker='o',markersize=10)
plt.title('virat vs rohit record')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='solid',linewidth=4,label='virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='dashdot',linewidth=4,label='rohit')
plt.title('virat vs rohit record')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend(loc='upper right')
```

```python
# multiple bar charts
df=pd.read_csv('batsman_season_record.csv')
df
```

```python
# wrong way of multiple bar chart
plt.bar(df['batsman'],df['2015'],color='red')
plt.bar(df['batsman'],df['2016'],color='blue')
plt.bar(df['batsman'],df['2017'],color='green')
```

```python
# multiple plot
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.2,label='2015')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,label='2016')
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width=0.2,label='2017')

plt.xticks(np.arange(df.shape[0]),df['batsman'])

plt.xlabel('batsman_name')
plt.ylabel('run_scored')
plt.title('bastman run record')
plt.legend()
plt.show
```

```python
# stascked bar chart
plt.bar(df['batsman'],df['2015'],label='2015')
plt.bar(df['batsman'],df['2016'],bottom=df['2015'],label='2016')
plt.bar(df['batsman'],df['2017'],bottom=(df['2015']+df['2016']),label='2017')
plt.legend()
plt.show()
```

```python
df=pd.read_csv('batsman_season_record.csv')
df
```

```python
plt.pie(df['2016'],labels=df['batsman'])
plt.show()
```

```python
# to show percentage
plt.pie(df['2016'],labels=df['batsman'],autopct='%0.1f%%')
plt.show()
```

```python
# percentage and color
plt.pie(df['2016'],labels=df['batsman'],autopct='%0.1f%%',colors=['red','blue','green','pink','yellow'])
plt.show()
```

```python
# explode and shadow

plt.pie(df['2016'],labels=df['batsman'],autopct='%0.3f%%',colors=['red','blue','green','pink','yellow'],explode=[0.1,0,0,0,0.4],shadow=True)
plt.show()
```
