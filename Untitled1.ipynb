```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('iris.csv')
```

```python
batters=pd.read_csv('batter.csv')
```

```python
batters.head()
```

```python
sample_df=batters.head(100).sample(25,random_state=5)
```

```python
sample_df.head()
sample_df.shape[0]
```

```python
# Annotation
plt.figure(figsize=(14,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],c=sample_df['runs'],s=sample_df['runs'])
for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
plt.colorbar()    
```

```python
x=[1,2,3]
y=[5,6,7]
plt.scatter(x,y)
plt.text(1,5,'point1')
plt.text(2,6,'point2')
plt.text(3,7,'point3',fontdict={'size':12,'color':'red'})
```

```python
# horizental line and vertical line
plt.figure(figsize=(14,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],c=sample_df['runs'],s=sample_df['runs'])
plt.axhline(130,color='red')   #horizental line
plt.axhline(140,color='blue')
plt.axvline(30,color='green')  #green
for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
plt.colorbar()    
```

```python
# subplots
fig,ax=plt.subplots(figsize=(10,7))
ax.scatter(batters['avg'],batters['strike_rate'],marker='+')
ax.set_title("something")
ax.set_xlabel('avg')
ax.set_ylabel('strike_rate')
fig.show
```

```python
fig,ax=plt.subplots(nrows=2,ncols=1,figsize=(10,7),sharex=True)
ax[0].scatter(batters['avg'],batters['strike_rate'],color='red')
ax[1].scatter(batters['avg'],batters['runs'],marker='+')
ax[0].set_title('avg vs strike rate')
ax[0].set_xlabel('avg')

ax[0].set_ylabel('strike_rate')

ax[1].set_title('avg vs run')
ax[1].set_xlabel('avg')
ax[1].set_ylabel('run')
```

```python
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(10,10))
ax[0,0].scatter(batters['avg'],batters['strike_rate'],color='red')
ax[0,1].scatter(batters['avg'],batters['runs'],marker='+')
ax[1,0].hist(batters['runs'])
ax[1,1].hist(batters['avg'])
```

```python
fig=plt.figure(figsize=(10,10))

ax1=fig.add_subplot(2,2,1)
ax1.scatter(batters['avg'],batters['strike_rate'],color='red')

ax2=fig.add_subplot(2,2,2)
ax2.scatter(batters['avg'],batters['runs'],marker='+')

ax3=fig.add_subplot(2,2,3)
ax3.hist(batters['runs'])

ax4=fig.add_subplot(2,2,4)
ax4.hist(batters['avg'])
```

```python
# 3D scatter plot
fig=plt.figure()

ax=plt.subplot(projection='3d')
ax.scatter3D(batters['avg'],batters['runs'],batters['strike_rate'],color='red',marker='+')
ax.set_title('ipl batsman analysis')
ax.set_xlabel('avg')
ax.set_ylabel('runs')
ax.set_zlabel('strike rate')
```

```python
# 3D line plot
x=[0,1,5,15]
y=[0,10,13,10]
z=[0,13,20,18]

fig=plt.figure()

ax=plt.subplot(projection='3d')
ax.scatter3D(x,y,z,color='red',marker='o',s=[100,100,100,100])
ax.plot3D(x,y,z,color='green')
```

```python
# 3D surface plot
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
```

```python
xx,yy=np.meshgrid(x,y)
```

```python
xx.shape
```

```python
z=xx**2+yy**2
```

```python
fig=plt.figure()
ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
ax.set_title("something")
```

```python
z=np.sin(xx)+np.cos(yy)
fig=plt.figure()
ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
ax.set_title("something_extra")
```

```python
z=np.sin(xx**2)+np.cos(yy**2)
fig=plt.figure()
ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
ax.set_title("something_more extra")
```

```python
z=np.sin(xx**2)+np.log(yy**2)
fig=plt.figure()
ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
ax.set_title("something_more extra extra")
```

```python
# contour plot
z=xx**2+yy**2
fig=plt.figure()
ax=plt.subplot()
p=ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(p)
ax.set_title("something_more extra")
```

```python
# contour fill
z=xx**2+yy**2
fig=plt.figure()
ax=plt.subplot()
p=ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
ax.set_title("something_more extra")
```

```python
z=np.sin(xx)+np.cos(yy)
fig=plt.figure()
ax=plt.subplot()
p=ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
ax.set_title("something_more extra")
```

```python
# heatmap
delivery=pd.read_csv('IPL_Ball_by_Ball_2008_2022.csv')
delivery.head()
```

```python
temp_df=delivery[(delivery['ballnumber'].isin([1,2,3,4,5,6])) & (delivery['batsman_run']==6)]
```

```python
grid=temp_df.pivot_table(index='overs',columns='ballnumber',values='batsman_run',aggfunc='count')
```

```python
grid
```

```python
# heatmap
fig=plt.figure(figsize=(20,10))
plt.imshow(grid)
plt.yticks(delivery['overs'].unique(),list(range(1,21)))
plt.xticks(np.arange(0,6),list(range(1,7)))
plt.colorbar()
```
